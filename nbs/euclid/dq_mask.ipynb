```python
# | hide

from astropy.io import fits

from nicl.euclid.utilities import (
    default_data_path,
    get_dq_mask,
    get_nisp_dither,
    get_vis_dither,
)
```

## Produce DQ pixel mask examples

```python
path = default_data_path("Q1_R1")
filter = "NIR_J"
obs_id = 2681
dither = 1
detector = 11
fn = get_nisp_dither(obs_id, dither, filter, path)
sci_ext = f"DET{detector}.SCI"
dq_ext = f"DET{detector}.DQ"
dqhdr = fits.getheader(fn, extname=dq_ext)
dq_flags = {k.split("_")[-1]: v for k, v in dqhdr.items() if "MSK_FLAG" in k}
outpath = default_data_path("dq_mask_examples/NIR")
outpath.mkdir(parents=True, exist_ok=True)
```

```python
for k, v in dq_flags.items():
    mask = get_dq_mask(fn, dq_ext, v)
    mask = mask.astype(int)
    fits.writeto(outpath / f"{v:02d}_{k}.fits", mask, overwrite=True)
data = fits.getdata(fn, extname=sci_ext)
fits.writeto(outpath / "data.fits", data, overwrite=True)
```

```python
filter = "VIS"
obs_id = 2681
dither = "0-1"
detector = "3-1.G"
fn = get_vis_dither(obs_id, dither, path)
sci_ext = f"{detector}.SCI"
dq_ext = f"{detector}.FLG"
dqhdr = fits.getheader(fn, extname=dq_ext)
dq_flags = {
    "INVALID": 0,  #
    "HOT": 1,  # X
    "COLD": 2,  # X
    "SAT": 3,  # X
    "COSMIC": 4,  # X
    "GHOST": 5,  # X
    "QUADEDGE": 6,  #
    "BAD_COLUMN": 7,  # X
    "BAD_CLUSTER": 8,  # X
    "CR_REGION": 9,  #
    "LOWVAR_COL": 10,  #
    "OVRCOL": 12,  # X
    "CHARINJ": 15,  # X
    "SATXTALKGHOST": 17,  # X
    "STARSIGNAL": 18,  #
    "ADCMAX": 21,  # X
    "NO_DATA": 22,  # X
    "STITCHBLOCK": 23,  #
    "OBJECTS": 24,  #
}
outpath = default_data_path("dq_mask_examples/VIS")
outpath.mkdir(parents=True, exist_ok=True)
```

```python
for k, v in dq_flags.items():
    mask = get_dq_mask(fn, dq_ext, v)
    mask = mask.astype(int)
    fits.writeto(outpath / f"{v:02d}_{k}.fits", mask, overwrite=True)
data = fits.getdata(fn, extname=sci_ext)
fits.writeto(outpath / "data.fits", data, overwrite=True)
```
