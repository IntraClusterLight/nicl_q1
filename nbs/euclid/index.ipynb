# Euclid

> Low surface-brightness reduction and analysis pipeline for Euclid data

## Data access

To avoid needing to enter your ESA username and password when accessing the online archive, create a file called `~/.euclid_credentials` containing the following:
```
user: <your_esa_username>
password: <your_esa_password>
```

## Data structure

We want flexibility to run the functions individually and interactively, but also that they can be put together in a pipeline. The following rules are designed to help with that.

- All functions should take either the absolute path to any input files or an argument giving the path that will be searched recursively to find the required files.
- All functions should take an argument where they should output any files they produce.

For the pipeline we should assume the following folder structure. The euclid_q1 branch matches how the files are stored in the Euclid archive (except for RAW, which I have put in Q1_R1 for convenience), which is directly accessible from the Euclid DataLab. The processed branches may be refined. To avoid hardcoding the absolute path into the notebook examples, the environment variable `$EUCLID_DATA` will be used to refer to the euclid_data folder.

- euclid_data
    - Q1_R1
        - MER
            - *tile_index*
                - NIR
                    - *fits files*
                - VIS
                    - *fits files*
        - NIR
            - *observation_id*
                - *fits files*
        - RAW
            - *observation_id*
                - *fits files*

                - *fits files*

## Reduction

This will ultimately include the following elements:

1) Data access
   * given an ra, dec and radius, determine the images that include the target, as well as those surrounding in time and sky position that are used in the reduction
   * download the reduced, but not stacked, images for a given observation_id
   * calibrated frames, MER mosaics, and additional data, e.g. raw slew darks and SIR dispersed frames
2) Persistence
   * determine and apply a simple persistence correction
3) Flatfield
   * create a skyflat to test small-scale flat-fielding and calibrate a large-scale flat-fielding correction
4) PSF
   * create a large-scale PSF model
   * subtract bright stars from the image
5) Stack
   * combine multiple dithers, potentially for multiple observations_ids
6) Cirrus
   * identify/model/subtract cirrus
7) Background
   * model and subtract the large-scale background
   * estimate the scale-dependent systematic background uncertainty
   * create a statistical uncertainty image, accounting for pixel correlations
8) Cutout
   * cutout the cluster region from the full image
9) Analysis (see below)
   * ICL profile, flux, shape, etc.
11) Cleanup
    * delete (some of) the downloaded / intermediate images to save space

The following will be required in several of these elements:

* Masking
* Estimates of the background error

The code for each reduction step should primarily return images ready for use in the next step, but include an option to also save intermediate images to disk.
