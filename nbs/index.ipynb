# nicl

> New Nottingham ICL code

The aim of this package is to build a consistent and tested suite of code for analysing Intracluster Light within the Nottingham ICL group. Ultimately, this package -- or parts of it -- will be shared with collaborators and made public so that others can reproduce our published analyses.

## Installation

This software is still in active development, but is starting to be used within the Nottingham ICL group.


### Basic install
If you only want to use (not develop) the software, you can install it (and its dependencies) into an environment by cloning this repository, changing to the root repository folder, and running:
```
pip install .
```

You will also need working installations of SWarp and AutoProf on your path.

#### SWarp

If you are using a conda environment you can install SWarp with `conda install conda-forge::astromatic-swarp`. Alternatively, see the [SWarp homepage](https://www.astromatic.net/software/swarp/) for Linux binaries or to build from source.

#### AutoProf

While AutoProf is a Python package, it currently (as of May 2025) requires old versions of several key packages, e.g. photutils. To avoid the same requirements for nicl, AutoProf is not included as a dependency, but instead should be installed in a separate environment. To achieve this, using conda, you can do the following
```
conda create -y -n autoprof python=3.12
conda activate autoprof
pip install autoprof
AUTOPROF=`which autoprof`
conda deactivate
```

Then, activate the environment in which you have installed nicl, e.g. `conda activate icl`, and create a symlink the main AutoProf script in the binary folder, as follows:
```
ln -sf $AUTOPROF $CONDA_PREFIX/bin/
```

### Development install

To install the software for development, first create a clean Python (>= 3.12) environment. This could be a virtualenv or a conda environment (e.g. `conda create -n icl python=3.12`). Clone this repository and change to the root repository folder. With your environment activated (e.g. `conda activate icl`), install nicl and all its dependencies with:
```
pip install -e '.[dev]'
```
The `-e` here performs an 'editable' installation, i.e. the files in the repository are used directly, rather than being copied to a central location.

To run particular elements of the code you will also need to have SWarp and/or AutoProf installed; see the instructions above.

The core code, documentation, tests and applications are written in Jupyter notebooks. To faciliate this, the development of this software uses [nbdev](https://nbdev.fast.ai). To understand this system, you must familiarise yourself with the [introductory walkthrough](https://nbdev.fast.ai/tutorials/tutorial.html). 

To ensure the code committed to the repository is well-formatted and correctly cleaned and exported, we use some git commit hooks. To enable these run:
```
pre-commit install
```
Now, whenever you try to commit to this repo, a set of checks will run.

#### Committing changes to git

Start a commit by doing `git add your_new_or_updated_notebooks.ipynb`. Then run `git commit -m "Your informative log message"`. Typically the notebook will be reformatted, cleaned and exported, and the checks will fail. Next, do `git add your_new_or_updated_notebooks.ipynb` again, and also add any new files that have been created in the `nicl` folder. You might find `git add -u .` helpful, but be careful not to add any files you have changed but don't want to commit at this stage. Then run `git commit -m "Your informative log message"` again. You will likely need to go through this routine one more time, then the commit will succeed. If it doesn't, there may be a problem with your code, which you will need to fix before commiting. If you get desperate, you can skip the checks with `git commit -n`, then let Steven know so he can investigate.

## Structure

The notebooks are in the [`nbs` folder](https://github.com/IntraClusterLight/nicl/tree/main/nbs). These are automatically exported by nbdev to produce the library in the [`nicl` folder](https://github.com/IntraClusterLight/nicl/tree/main/nicl). The notebooks can use code defined in each other by importing the pip-installed library, i.e. `import nicl`.  The `nicl` folder can also contain additional python code that is not generated from the notebooks; this currently includes a modified version of ezgal.

The root [`nbs` folder](https://github.com/IntraClusterLight/nicl/tree/main/nbs) contains notebooks defining library functionality that should be of general use in analysing ICL in a variety of datasets. Within that folder are subfolders containing notebooks defining library functionality tailored to specific datasets, e.g. HSC and Euclid. Those folders can also contain notebooks that perform the steps for a particular analysis, making use of the functionality in the nicl library.
