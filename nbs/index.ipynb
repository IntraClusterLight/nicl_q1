# nicl_q1

> Nottingham ICL data processing, measurement and analysis package – Euclid Q1 version

The nicl package is a suite of code for studies of Intracluster Light, primarily intended for use by the Nottingham ICL group and collaborators. While we aim for many elements to be applicable to a range of datasets, a substantial section of the current code is focussed on Euclid data. This nicl_q1 repository contains a branch of our main (currently private) nicl repository, which captures the state of nicl development at the time we processed the Euclid Q1 data. The details of the processing and measurement approach are presented in Euclid Collaboration: Bamford et al. (2026). These measurements have been used in the following papers:

- Euclid Collaboration: Kolcu et al. (2026)
- Euclid Collaboration: Golden-Marx et al. (2026)
- Euclid Collaboration: Gully et al. (2026)

## Structure and caveats

The code is implemented in notebooks located in the ["nbs" folder](https://github.com/IntraClusterLight/nicl_q1/tree/main/nbs). The functions defined in these notebooks are automatically exported by [nbdev](https://nbdev.fast.ai) (v2) to produce the library in the ["nicl" folder](https://github.com/IntraClusterLight/nicl_q1/tree/main/nicl). The notebooks can therefore use code defined in each other by importing the pip-installed library, i.e. `import nicl`.  Many notebooks also contain examples illustrating the development and usage of the library functions. [Documentation](https://intraclusterlight.github.io/nicl_q1) is automatically generated from the notebooks. The "nicl" folder can also contain additional python code that is not generated from the notebooks; this currently includes a modified version of ezgal.

The root ["nbs" folder](https://github.com/IntraClusterLight/nicl/tree/main/nbs) contains notebooks defining library functionality that should be of general use in analysing ICL in a variety of datasets. Within that folder are subfolders containing notebooks defining library functionality tailored to specific datasets, e.g. Euclid. The notebooks can also perform analysis steps, but not contribute functionality to the nicl library. Many of the plots in the Euclid Collaboration: Bamford et al. (2026) paper are created in "nbs/euclid/q1/data_processing.ipynb".

This repository is made publicly available for transparency and reproducibility, but it is a snapshot of work in progress and not intended as a ready-to-use package. This will be improved in versions accompanying future papers, e.g. based on the Euclid DR1 release. If you are trying to reproduce the papers listed above, then the following scripts give a outline of the steps required. If you encounter any difficulties, please reach out to the [Nottingham ICL team](mailto:icl-group@nottingham.ac.uk) and we will do our best to help.

The files "scripts/v0.7/run_pipeline_nir.py" and "run_pipeline_vis.py" download the necessary data, create the skyflats, perform the NIR persistence correction, produce combined images for each observation then measure some statistics. In practice, for large numbers of observations if it better to run each step separately, e.g. "scripts/get_all_nir.py", "scripts/create_coarse_data.py", "scripts/do_persistence_correction.py". Furthermore, in our Q1 papers we use combined images centred on target clusters, created with "scripts/stack_Q1.py". The ICL measurements are performed by "scripts/run_measure.py". Typically each of these scripts was not run directly, but submitted to a SLURM queue on an HPC system (ada). The corresponding scripts are in "scripts/ada/".

The statistics presented in the Euclid Collaboration: Bamford et al. (2026) paper are measured on "skypatches",created by the scripts in "scripts/ada/skypatches" and measured by those in "scripts/ada/measure_noise". The creation and measurement of the mock clusters is described in "scripts/ada/mock_clusters", while measurements of the simulated (Hydrangea) clusters were performed using "scripts/ada/simulated_clusters". Ablation tests to illustrate the effect of each processing step were performed using the scripts in "scripts/ada/illustration".

## Installation

### Basic install
If you only want to use (not develop) the software, you can install it (and its dependencies) into an environment by cloning this repository, changing to the root repository folder, and running:
```
pip install .
```

You will also need working installations of SWarp and AutoProf on your path.

#### SWarp

If you are using a conda environment you can install SWarp with `conda install conda-forge::astromatic-swarp`. Alternatively, see the [SWarp homepage](https://www.astromatic.net/software/swarp/) for Linux binaries or to build from source.

#### AutoProf

While AutoProf is a Python package, it currently (as of July 2026) requires old versions of several key packages, e.g. photutils. To avoid the same requirements for nicl, AutoProf is not included as a dependency, but instead should be installed in a separate environment. To achieve this, using conda, you can do the following
```
conda create -y -n autoprof python=3.12
conda activate autoprof
pip install autoprof
AUTOPROF=`which autoprof`
conda deactivate
```

Then, activate the environment in which you have installed nicl, e.g. `conda activate icl`, and create a symlink the main AutoProf script in the binary folder, as follows:
```
ln -sf $AUTOPROF $CONDA_PREFIX/bin/
```

### Development install

We do not recommend developing against this repository, since significant further work has already taken place on the main (private) repository. If you have feedback or would like to contribute, either create an issue or contact the team [Nottingham ICL team](mailto:icl-group@nottingham.ac.uk).

To install the software for development, first create a clean Python (>= 3.13) environment. This could be a virtualenv or a conda environment (e.g. `conda create -n icl python=3.13`). Clone this repository and change to the root repository folder. With your environment activated (e.g. `conda activate icl`), install nicl and all its dependencies with:
```
pip install -e '.[dev]'
```
The `-e` here performs an 'editable' installation, i.e. the files in the repository are used directly, rather than being copied to a central location.

To run particular elements of the code you will also need to have SWarp and/or AutoProf installed; see the instructions above.

The core code, documentation, tests and applications are written in Jupyter notebooks. To faciliate this, the development of this software uses [nbdev](https://nbdev.fast.ai). To understand this system, you must familiarise yourself with the [introductory walkthrough](https://nbdev.fast.ai/tutorials/tutorial.html). 

To ensure the code committed to the repository is well-formatted and correctly cleaned and exported, we use some git commit hooks. To enable these run:
```
pre-commit install
```
Now, whenever you try to commit to this repo, a set of checks will run.
